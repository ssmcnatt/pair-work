In this assignment, we will learn about other optimization algorithms. We will create a neural network and try out the different optimization algorithms and compare their performance.

```python
import numpy as np
import pandas as pd

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.optimizers import Adagrad, RMSprop, Adam

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import StandardScaler
```

In this assignment, we will be using the cancer data that we have worked with in previous lessons. The pre-processed data is loaded below.

```python
cancer = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/cancer_processed.csv')
```

```python
cancer.head()
```

As you may recall, diagnosis is the target variable. One hot encode the diagnosis column.

```python
# Answer below:
cancer['diagnosis'] = pd.get_dummies(cancer['diagnosis'], drop_first=True)
cancer.head()
```

Split the data into train and test with 20% of the data in test.

```python
X = cancer.drop(columns=['diagnosis'])
y = cancer['diagnosis']
```

```python
# Answer below
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=66)
```

Scale all other variables using the standard scaler.

```python
# Answer below:
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

Generate a sequential model consisting of 5 layers. The layers should be of size 128, 64, 32, 32, 1. Use the appropriate activation for the output layer based on the type of prediction algorithm we are producing.

```python
# Answer below
model = Sequential()

# first dense layer
model.add(Dense(128, input_dim=X_train.shape[1], activation='relu'))

# second dense layer
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(32, activation='relu'))

# output dense layer
model.add(Dense(1, activation='sigmoid')) 
```

Initialize a SGD optimizer with learning rate 0.05 and momentum 0.9.

```python
# Answer below:
sgd = SGD(learning_rate=0.05, momentum=0.9)
```

Compile and fit the model using the appropriate loss function and metric and use the optimizers defined above.

batch size = 100, epochs = 200

```python
# Answer below:
model.compile(optimizer=sgd, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, validation_data=(X_test, y_test), batch_size=100, epochs=200, verbose=1)
```

Define the RMSprop optimizer with a learning rate of 0.05.

```python
# Answer below:
rms_model = Sequential()

# first dense layer
rms_model.add(Dense(128, input_dim=X_train.shape[1], activation='relu'))

# second dense layer
rms_model.add(Dense(64, activation='relu'))
rms_model.add(Dense(32, activation='relu'))
rms_model.add(Dense(32, activation='relu'))

# output dense layer
rms_model.add(Dense(1, activation='sigmoid'))
```

Compile and fit the model using the optimizer defined above. What do you notice about the accuracy?

```python
# Answer below:
rms = RMSprop(learning_rate=0.05)
rms_model.compile(optimizer=rms, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
rms_model.fit(X_train, y_train, validation_data=(X_test, y_test), batch_size=100, epochs=200, verbose=1)
```

Define the Adam optimizer with learning rate 0.01

```python
# Answer below:
adam_model = Sequential()

# first dense layer
adam_model.add(Dense(128, input_dim=X_train.shape[1], activation='relu'))

# second dense layer
adam_model.add(Dense(64, activation='relu'))
adam_model.add(Dense(32, activation='relu'))
adam_model.add(Dense(32, activation='relu'))

# output dense layer
adam_model.add(Dense(1, activation='sigmoid'))
```

```python
adam = Adam(learning_rate=0.01)
```

Compile and fit the model using the optimizer defined above. How does the peformance differ with this optimizer?

```python
# Answer below:
adam_model.compile(optimizer=adam, loss='binary_crossentropy', metrics=['accuracy'])
adam_model.fit(X_train, y_train, validation_data=(X_test, y_test), batch_size=100, epochs=200, verbose=1)
```
